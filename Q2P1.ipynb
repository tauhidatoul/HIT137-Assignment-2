```python
pixel = []
coordinates = []
for a in range(0, 1000):
    for b in range(0,1500):
        rgb = px[a, b]
        coordinates.append([a,b])
    
```

```python
import pandas as pd

column = ['Coordinates', 'RGB']
df = pd.DataFrame(zip(coordinates, pix_val), columns=column)
```

```python
list_of_lists = [list(t) for t in pix_val]
```

```python
list_of_lists
```

```python
import time
current_time = int(time.time())
generated_number = (current_time%100)+50
if generated_number % 2 ==0:
    generated_number += 10
print(generated_number)
```

```python
list_with_added_number = [[element + generated_number for element in sublist] for sublist in list_of_lists]
```

```python
list_with_added_number
```

```python
list_of_tuples_after_addition = [tuple(sublist) for sublist in list_with_added_number]
```
